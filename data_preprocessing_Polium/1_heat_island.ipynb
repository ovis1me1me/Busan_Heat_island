```python
import folium
from folium import plugins
from folium import Marker
from folium.plugins import HeatMap
from folium.plugins import MarkerCluster
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/output.csv', encoding='cp949')
df['구'] = df['address'].str.split(' ').str[1]
df_sorted=df.sort_values(by='구')
df
```

```python
df = df.dropna(subset=['latitude', 'longitude'])

latitude_series = df['latitude']
longitude_series = df['longitude']
address_series = df['address']

# 처음 나타나는 지도 위치와 확대의 정도
m = folium.Map(location=[35.2, 129], zoom_start=11)

# 위도, 경도, 주소를 lat, lon, loc로 받아 표시
for lat, lon, loc in zip(latitude_series, longitude_series, address_series):
  popup_text = folium.Popup(loc, max_width=200)  # 마커를 누르면 나타나는 주소 지정
  folium.Marker(location=[lat, lon], popup=popup_text).add_to(m)

  # 위치에 빨간 원을 만들어 겹친부분 확인
  folium.Circle(location=[lat, lon], radius=200, color='red', fill=True, fill_color='red').add_to(m)
# 마우스 위치에 따른 좌표 표시
plugins.MousePosition().add_to(m)
m
```

```python
m1 = folium.Map(location=[35.2, 129], zoom_start=11)

# 지도를 축소하면 주변 지역을 묶어 개수 분포 확인
mc = MarkerCluster()
for _, row in df.iterrows():
  popup = folium.Popup(row['address'], max_width=200)
  folium.Marker(location=[row['latitude'], row['longitude']], popup=popup).add_to(mc)
plugins.MousePosition().add_to(m1)
mc.add_to(m1)
m1
```

```python
m3=folium.Map(location=[35.2, 129], zoom_start=11)

# 구별로 마커 색 지정
color_mapping = {
    '금정구':'red',
    '해운대구':'orange',
    '동구':'green',
    '강서구':'blue',
    '사상구':'purple',
}

for _, row in df.iterrows():
    area = row['구']
    color = color_mapping.get(area)

    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=folium.Popup(row['address'], max_width=200),
        icon=folium.Icon(color=color)
    ).add_to(m3)

plugins.MousePosition().add_to(m3)

m3
```

```python
# HeatMap
m2 = folium.Map(location=[35.2, 129], zoom_start=11)
HeatMap(
    data=df[['latitude', 'longitude']], # heatmap을 그릴 위치 표시
    radius=20, # 원 크기
).add_to(m2)
plugins.MousePosition().add_to(m2)
m2
```
